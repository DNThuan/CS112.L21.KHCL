# 4. Đánh giá độ phức tạp
-Hàm list.sort() trong python là một hàm sort kết hợp giữa Insertion Sort-O(n^2) và Merge Sort-O(nlog(n)) nên độ phức tạp là O(nlog(2n))
- Vậy độ phức tạp bài toán là O(nlog(2n)+n) = O(nlog(2n))

# 5.Programming

````python
s=input().strip()
a=list(map(int,[i for i in s]))

s=sum(a)
l=len(a)
a.sort()
# 2 5 8 
if s%3==2: 
    i=0
    while i<l:
        if a[i]%3==2:
            a.pop(i)
            break
        i+=1
    if i==l:
        c=2
        j=0
        while c>0:
            if a[j]%3==1:
                a.pop(j)
                c-=1
            else:
                j+=1
#1 4 7 
elif s%3==1:
    i=0
    while i<l:
        if a[i]%3==1:
            a.pop(i)
            break
        i+=1
    if i==l:
        c=2
        j=0
        while c>0:
            if a[j]%3==2:
                a.pop(j)
                c-=1
            else:
                j+=1
#0 3 6 9 
else:
    pass
a.sort(reverse=True)

for i in a:
    print(i,end='')
````
