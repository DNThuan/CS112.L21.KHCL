# 4) Độ phức tạp
- Tốt nhất O(1)
- Tệ nhất O(n)
- Trung bình O(n)

# 5) Code

````python
a,k,b,m,n = list(map(int,input().split()))

day_T=n//(a+b)
tree_T= day_T*(a+b)-a*(day_T//k)-b*(day_T//m)
other=n-tree_T

while other>=(a+b):
    day_T=day_T+other//(a+b)
    tree_T=day_T*(a+b)-a*(day_T//k)-b*(day_T//m)
    other=n-tree_T

while tree_T <n:
    day_T+=1
    tree_T+=(a+b)
    if day_T%k==0:
        tree_T-=a
    if day_T%m==0:
        tree_T-=b  

print(day_T)
````
