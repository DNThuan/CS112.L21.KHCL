# 2) Pattern Recognition

# 3) Algrorithm

Thuật toán của em giải quyết bài toán này rất đơn giản: duyệt từ đầu dãy kiểm tra kèm theo các điều kiện, khi thích hợp sẽ thay thế cho ra kết quả.

##Kết quả bài toán (m) cần phải thỏa mãn 3 điều kiện sau:

1. Chỉ khác n duy nhất một chữ số
  - Ta sẽ duyệt, kiểm tra từng số duy nhất, ứng với mỗi trường hợp ta sẽ quyết định bước tiếp theo là bỏ qua hay biến đổi nó thành kết quả.
2. Là số lớn nhất
  - Để đám bảo kết quả (m) là số lớn nhất, ta kiểm duyệt, biến đổi từ trái sang phải.
3. Là số chia hết cho 3
  - Khi chọn được mục tiêu cần biến đổi, kiểm tra toàn cục và quyết định chọn số nào để thay thế và đảm bảo kết quả luôn chia hết cho 3.

- Ta quan tâm tới 3 số 7, 8, 9 là 3 số lớn nhất ứng với 3 trường hợp chia 3 dư 1, 2, 0. Ta lần lượt kiểm tra 3 số này trước.

  - Số 9 là số lớn nhất nên bỏ qua.
  - Số 8, kiểm tra xem có thể thay bằng số 9 được hay không, nếu không bỏ qua
  - Số 7, kiểm tra xem có thể thay bằng số 8 hoặc 9 hay không, nếu không bỏ qua
- Đổi với các số còn lại luôn thay thể được bởi một trong các số 7,8 hoặc 9
- Trường hợp duyệt hết dãy mà chưa tìm được cách thay thế tức dãy chứa các số lớn 7,8,9 thì ta thay đổi số cuối cùng trong dãy bằng cách trừ đi 1,2,3 ứng với số dư n%3 là 1,2,0.







# 4) Độ phức tạp
- Tốt nhất: O(1) xảy ra khi số thay đổi là đầu dãy
- Xấu nhất: O(n) xảy ra khi duyệt hết dãy
- Trung bình O(n)

# 5) Code

```python
def SoNguyenMoi(a):
    a=[int(x) for x in a]
    s=sum(a)
    i=0
    while i<len(a):
        temp=a[i]
        if temp==9:
            i+=1
        elif temp==8:
            if (s+1) % 3==0:
                a[i]=9
                break
            else:
                i+=1
        elif temp==7:
            if (s+1) %3==0:
                a[i]=8
                break
            if (s+2) % 3==0:
                a[i]=9
                break
            else:
                i+=1
        else:
            a[i]=0
            s=sum(a)
            if s%3==0:
                a[i]=9
                break
            elif s%3==2:
                a[i]=7
                break
            elif s%3==1:
                a[i]=8
                break
        if i==len(a):
            if s%3==1:
                a[-1]-=1
                break
            elif s%3==2:
                a[-1]-=2
                break
            else:
                a[-1]-=3
    s=''
    for i in a:
        s+=str(i)
    return s

a=input()
print(SoNguyenMoi(a))
```
