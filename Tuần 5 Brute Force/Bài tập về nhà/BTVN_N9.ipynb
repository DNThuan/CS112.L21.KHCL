```python
#Bài 1: 3 phần tử có tích lớn nhất
import sys

arr = list(map(int,input().split()))
n = len(arr)
max_product = -(sys.maxsize)
for i in range(0, n - 2):
  for j in range(i + 1, n - 1):
    for k in range(j + 1, n):
      if (max_product < arr[i] * arr[j] * arr[k]):
        max_product = arr[i] * arr[j] * arr[k]

print(max_product)
```

```python
#Bài 2: Số phong phú

import math

def getSum(n) :
    sum = 0
    i = 1
    while i <= (math.sqrt(n)) :
        if n%i == 0 :
            if n/i == i :
                sum = sum + i
            else : 
                sum = sum + i
                sum = sum + (n / i )
        i = i + 1
    sum = sum - n
    return sum
 
def checkAbundant(n) :
    if (getSum(n) > n) :
        return 1
    else :
        return 0
         
count = 0
a = list(map(int,input().split()))
for i in range(a[0],a[1]):
  if(checkAbundant(i) == 1) :
    count+=1
print(count)
```
