# 1) Abstraction
Xuất ra số lần biến đổi từ phân số a/b để trở thành phân số c/d. Nếu không có cách thì xuất ra 0

# 2) Pattern Recognition
- Tối ưu hóa và tính toán các trường hợp đặt biệt

# 3) Algorithm
- Liên tục biến đổi (cộng tử mẫu cho 1 và rút gọn) kèm theo các điều kiện cho trường hợp không thể tồn tại số lần biến đổi từ a/b sang c/d như:

  + a/b: Tử lớn (nhỏ) hơn mẫu và c/d: Tử nhỏ (lớn) hơn tử.
  + abs(a-b) = 1 và abs(c-d) != 1
  + abs(a-b) = abs(c-d) =1 và a>c

# 4) Complexity
- Hàm tìm UCLN dựa trên Euclidean algorithm by division nên có độ phức thời gian là O(log(a+b)).
- Dể dàng nhập thấy phân số a/b luôn trở thành dạng abs(a-b)=1, lúc này chúng ta dể dạng đưa ra kết luận nếu phân số c/d không cùng dạng với a/b. Trường hợp vòng while thực hiện nhiều lần nhất khi abs(a-b) = abs(c-d) = 1. Vậy độ phức tạp thời gian là: O(k.log(a+b)+c-a) với k là số lần rút gọn trước khi về dạng abs(a-b)=1.


# 5) Programming



```python
a=int(input())
b=int(input())
c=int(input())
d=int(input())

def UCLN(a,b):
    while a!=0:
        a,b=b%a,a
    return b

def KetQua(a,b,c,d):
    count =0
    while (a!=c) or (b!=d):
        if abs(a-b)==1:           
            if abs(c-d)==1:
                if (a>b and c>d) or (a<b and c<d):
                    if a>c:
                        return 0
                else:
                    return 0
            else: 
                return 0
        a+=1
        b+=1
        u=UCLN(a,b)
        if u !=1:
            a=a//u
            b=b//u
        count+=1
    return count            

print(KetQua(a,b,c,d))
```
