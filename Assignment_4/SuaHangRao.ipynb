# Problem: Sửa hàng rào
image.png

# 1) Abstraction
Tìm trong và đóng lần lượt các tấm gỗ sao cho sau khi đóng thì độ cao của tấm gỗ thấp nhất trong hàng rào là cao nhất.

# 2) Pattern Recognition
Chia để trị
- Phương pháp sử dụng là tìm kiếm nhị phân

# 3) Algorithm designed

- Tìm chiều cao tối đa
```
Trong left < right thì:
  mid = (left + right)/2 là tấm gỗ được chọn.
  Nếu các tấm gỗ được chọn nhỏ hơn mid thì
    left =mid+1 tức giá trị tối ưu nằm bên phải mảng
  Ngược lại
    right=mid là khoảng chứa giá trị tối ưu (gọi là h_new)


- Chọn thanh gỗ
```
Duyệt qua các tấm gỗ trong hàng rào
    Nếu tấm gỗ thứ i < h_new thì
      Duyệt tất cả các tấm gỗ chuẩn bị đóng:
        Nếu tổng tấm gỗ i và tấm chuẩn bị đóng > h_new thì
          Đánh dấu lại và đi đến tấm gỗ tiếp theo trên hàng rào.


# 4) Độ phức tạp 
- Tìm kiếm nhị phân O(logn)
- Độ phức tạp của hàm check là: O(n.m) khi duyệt tất cả các trường hợp
- Tổng độ phức tạp khi tìm h_new là O(n.m.logn)
- Tìm kiếm và xuất thanh được chọn đóng ra tốn chi phí O(n.m)
**Vậy tổng độ phức tạp cho thuật toán là O(n.m.logn)*

# 5) Programming
Do độ phức tạp của thuật toán xây dựng còn quá lớn nên chương trình đã không thể vượt qua một số test case lớn.

````python

n = int(input())
arr_n = list(map(int, input().split()))
m = int(input())
arr_m = list(map(int, input().split()))
min_arr_n = min(arr_n)
max_arr_m = max(arr_m)
max_mid = min_arr_n + max_arr_m 
min_mid = min_arr_n

def kiemtraGo(mid):
    j = 0
    for i in range(n):
        if arr_n[i] < mid:
            T = False
            while j < m:
                if arr_m[j] + arr_n[i] >= mid:
                    T = True
                    j += 1
                    break
                j += 1
            if T == False:
                return False
    return True

def Tim_h_new():
    l = min_mid
    r = max_mid
    while(l <= r):
        mid = (l + r)//2
        if (kiemtraGo(mid)):
            l = mid + 1
        else:
            r = mid - 1
    return r

def XuatKQ(h):
    j = 0
    KQ = []
    for i in range(n):
        if arr_n[i] < h:
            while j < m:
                if arr_m[j] + arr_n[i] >= h:
                    j += 1
                    break
                j+= 1
            KQ.append((i + 1, j))
    print(h, len(KQ))
    for x in KQ:
        print(x[0], x[1])


XuatKQ(Tim_h_new())
````
